### 1. Maximun product of Two Elements in an Array

```python
from typing import List
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        max1=0
        for i in range(len(nums)-1):
            for j in range(i+1,len(nums)):
                value=(nums[i]-1)*(nums[j]-1)
                if max1<value:
                    max1=value
        return max1
ans=Solution()
ans.maxProduct([3,4,5,2])
```

### 2. Count Number of Teams

```python
from typing import List
class Solution:
    def numTeams(self, rating: List[int]) -> int:
        high=[0]*len(rating)
        low=[0]*len(rating)
        c=0
        for i in range(len(rating)):
            for j in range(i+1,len(rating)):
                if rating[j]>rating[i]:
                    high[i]+=1
                elif rating[j]<rating[i]:
                    low[i]+=1
        for i in range(len(rating)):
            for j in range(i+1,len(rating)):
                if rating[j]>rating[i]:
                    c+=high[j]
                elif rating[j]<rating[i]:
                    c+=low[j]
        return c
ans=Solution()
ans.numTeams([2,5,3,4,1])
```

### 3.Number of Students Doing Homework at a Given Time

```python
class Solution:
    def busyStudent(self, startTime: List[int], endTime: List[int], queryTime: int) -> int:
        c=0
        for i in range(len(startTime)):
            if startTime[i]<=queryTime<=endTime[i]:
                c+=1
        return c
ans=Solution()
ans.busyStudent([1,2,3],[3,2,7],4)
```

### 4.Number of Steps to Reduce a Number to Zero

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        a=True
        c=0
        while num!=0:
            if num%2==0:
                num=num//2
                c+=1
            else:
                num=num-1
                c+=1
        return c
ans=Solution()
ans.numberOfSteps(14)
```

### 5. Count Bits

```python
from typing import List
class Solution:
    def countBits(self, n: int) -> List[int]:
        lst=[]
        for i in range(n+1):
            binary=bin(i)
            binary=binary[2:]
            if "1" in binary:
                num=binary.count("1")
                lst.append(num)
            else:
                lst.append(0)
        return lst
ans=Solution()
ans.countBits(2)
```
